```python
import csv
from datetime import datetime
from matplotlib import pyplot as plt
filename='data/temperature_2014.csv'
with open(filename) as f:
    reader=csv.reader(f)
    heade=next(reader)
    dates,highs,lows=[],[],[] #dates=[];highs=[];lows=[]
    for row in reader:
        try:
                                   #문자열을 날짜로  Y:년도 4자리 y:년도 2자리
            current_date=datetime.strptime(row[0],'%Y-%m-%d')
            high=int(row[1])
            low=int(row[3])
        except ValueError:
            print(current_date,'Missing date')
        else:
            dates.append(current_date)
            highs.append(high)
            lows.append(low)
            
#print(len(dates),len(highs),len(lows))

                #해상도  #크기
fig=plt.figure(dpi=128,figsize=(10,6))
                       #색상  #불투명도
plt.plot(dates,highs,c='red',alpha=0.5)
plt.plot(dates,lows,c='blue',alpha=0.5)
                                   #채우는 색 
plt.fill_between(dates,highs,lows,facecolor='blue',alpha=0.1)

plt.title('온도 통계',fontsize=20)
plt.xlabel('기간',fontsize=16)
plt.ylabel('온도',fontsize=16)
plt.tick_params(axis='both',labelsize=16)
                                                   #여백 
plt.savefig('data/images/bar_plot.png',dpi=400, bbox_nches='tight')
plt.show()
```

```python
df=pd.read_csv('data/temperature_2014.csv')
#df1=df.dropna(how='any')
df1=df[['PST','Max TemperatureF','Min TemperatureF']]
df1=df1.dropna(how='any')
dates=df1['PST'].astype('datetime64[D]') #datetime64[ns]
highs=df1['Max TemperatureF']
lows=df1['Min TemperatureF']

fig=plt.figure(dpi=128,figsize=(10,6))
                       #색상  #불투명도
plt.plot(dates,highs,c='red',alpha=0.5)
plt.plot(dates,lows,c='blue',alpha=0.5)
                                   #채우는 색 
plt.fill_between(dates,highs,lows,facecolor='blue',alpha=0.1)

plt.title('온도 통계',fontsize=20)
plt.xlabel('기간',fontsize=16)
plt.ylabel('온도',fontsize=16)
plt.tick_params(axis='both',labelsize=16)
                                                   #여백 
plt.savefig('data/images/bar_plot_pd.png',dpi=400, bbox_nches='tight')
plt.show()   
```
