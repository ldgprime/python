```python
client_id="pEKjmsHsctADzAEJ3hZ6"
client_secret="6oMYODyCQe"

#[CODE 1]
def get_request_url(url):
    
    req = urllib.request.Request(url)
    req.add_header("X-Naver-Client-Id", client_id)
    req.add_header("X-Naver-Client-Secret", client_secret)
    try: 
        response = urllib.request.urlopen(req)
        if response.getcode() == 200:
            print ("[%s] Url Request Success" % datetime.datetime.now())
            return response.read().decode('utf-8')
    except Exception as e:
        print(e)
        print("[%s] Error for URL : %s" % (datetime.datetime.now(), url))
        return None
```

```python
#[CODE 2]
def getNaverSearchResult(sNode, search_text, page_start, display):
    
    base = "https://openapi.naver.com/v1/search"
    node = "/%s.json" % sNode
    parameters = "?query=%s&start=%s&display=%s" % (urllib.parse.quote(search_text), page_start, display)
    #parameters = "?query=%s" % urllib.parse.quote(search_text)
#     parameters= "query=코로나&start=1&display=100"
    url = base + node + parameters
    
    retData = get_request_url(url)
    
    if (retData == None):
        return None
    else:
        return json.loads(retData)
```

```python
#[CODE 3]
def getPostData(post,jsonResult):
    
    title = post['title']
    description = post['description']
    org_link = post['originallink']
    link = post['link']

    #Tue, 14 Feb 2017 18:46:00 +0900
    pDate = datetime.datetime.strptime(post['pubDate'], '%a, %d %b %Y %H:%M:%S +0900')
    pDate = pDate.strftime('%Y-%m-%d %H:%M:%S')
    
    jsonResult.append({'title':title, 'description': description,
                    'org_link':org_link, 'link': link, 
                    'pDate':pDate})
    return
    
```

```python
#[CODE4]
def getDataFrame(jsonSearch,search_text):
    titles=[];descriptions=[];org_links=[];links=[];pDates=[]
    for row in jsonSearch['items']:
        titles.append(row['title'].strip())
        descriptions.append(row['description'].strip())
        org_links.append(row['originallink'])
        links.append(row['link'])
        #%a : 영어로 된 요일문자열, %d : 2자리 날짜, %b:영어로된 월 문자열
        pDate = datetime.datetime.strptime(row['pubDate'],
                               '%a, %d %b %Y %H:%M:%S +0900')
        pDate = pDate.strftime('%Y-%m-%d %H:%M:%S')
        pDates.append(pDate)
    
    df=pd.DataFrame({'title':titles,'description':descriptions,
                'org_link':org_links,'link':links,'pDate':pDates})
    print(df)
    df.to_csv('%s_naver.csv'%(search_text), mode='a',index=False)
    print ('%s_naver.csv SAVED' % (search_text))
    return df
```
