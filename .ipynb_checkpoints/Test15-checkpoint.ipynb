```python
import sqlite3
print(sqlite3.version)
print(sqlite3.sqlite_version)
```

```python
def create_table():
    #try except구문으로 해줘야함
    #실행 중 오류 발생시 db 락이 걸림
    #real 8byte 부동소수
    #sqlite 자료형: integer(boolean=0 or 1),real,text(date:'YYYY-MM-DD'), blob 
    try:
        conn=sqlite3.connect('my_books1.db')
        cursor=conn.cursor()
        sql="""create table if not exists books1(
        title text,
        published_date text,
        publisher text,
        pages interger,
        recommend integer
        )"""
        cursor.execute(sql)
    except:
        pass  
    finally:
        conn.commit()
        conn.close()
```

```python
 def insert():
    try:    
        conn=sqlite3.connect('my_books1.db')
        cursor=conn.cursor()
        sql="insert into books1 values('java','2019-01-05','길벗',500,10)"
        cursor.execute(sql)
        sql='insert into books1 values(?,?,?,?,?)'
    #     tuple(list)
                              #튜플형태로 넣어줌   
        cursor.execute(sql,('Python','2020-01-01','한빛',600,20))
        items={
             ('빅데이터','2019-02-04','영진',365,5),
             ('안드로이드','2018-02-03','인피니티북스',560,20),
             ('Spring','2020-02-04','삼성',600,15)
         }
        cursor.executemany(sql,items)
    except:
        pass

    finally:
        conn.commit()
        conn.close()
```

```python
def all_books():
    try:
        conn=sqlite3.connect('my_books1.db')
        cursor=conn.cursor()
        sql='select * from books1'
        cursor.execute(sql)
        books=cursor.fetchall()
        #데이터가 바뀌지 않기 때문에 commit 필요 x
    except:
        pass
    else:
        print(type(books))
        print(books)
    finally:
        conn.close()
    
```

```python
create_table()
insert()
all_books()
```

```python
def some_books(num):
    try:
        conn=sqlite3.connect('my_books1.db')
        cursor=conn.cursor()
        sql='select * from books1'
        cursor.execute(sql)
        books=cursor.fetchmany(num)
        #데이터가 바뀌지 않기 때문에 commit 필요 x
    except:
        pass
    else:
        print(type(books))
        print(books)
    finally:
        conn.close()
    
```

```python
def one_books():
    try:
        conn=sqlite3.connect('my_books1.db')
        cursor=conn.cursor()
        sql='select * from books1'
        cursor.execute(sql)
        books=cursor.fetchone()
        #데이터가 바뀌지 않기 때문에 commit 필요 x
    except:
        pass
    else:
        print(type(books))
        print(books)
    finally:
        conn.close()
```

```python
 def insert_movie(row):
    try:    
        conn=sqlite3.connect('movies.db')
        cursor=conn.cursor()   
        sql='insert into movies values(?,?,?,?)'  
        cursor.execute(sql,row)    
    except:
        print(error)
    finally:
        conn.commit()
        conn.close()
```
