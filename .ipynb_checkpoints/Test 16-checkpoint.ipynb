```python
import sqlite3
print(sqlite3.version)
print(sqlite3.sqlite_version)
```

```python
def create_table_movies():
    #try except구문으로 해줘야함
    #실행 중 오류 발생시 db 락이 걸림
    
    try:
        conn=sqlite3.connect('my_movies.db')
        cursor=conn.cursor()
        
        sql="""create table if not exists movies(
        id real,       
        title text,
        genres text,
        points real
        )"""
        cursor.execute(sql)
    except:
        print(error)
    finally:
        conn.commit()
        conn.close()    
```

```python
def all_movies():
    try:
        conn=sqlite3.connect('my_movies.db')
        cursor=conn.cursor()
        sql='select * from movies'
        cursor.execute(sql)
        books=cursor.fetchall()
        #데이터가 바뀌지 않기 때문에 commit 필요 x
    except:
        pass
    else:
        print(type(books))
        print(books)
    finally:
        conn.close()
```

```python
def some_movies(num):
    try:
        conn=sqlite3.connect('my_movies.db')
        cursor=conn.cursor()
        sql='select * from movies'
        cursor.execute(sql)
        movies=cursor.fetchmany(num)
        #데이터가 바뀌지 않기 때문에 commit 필요 x
    except:
        pass
    else:
        for movie in movies:
            print(movie)
    finally:
        conn.close()
    
```

```python
create_table_movies()
insert_movie()
all_movies()
```

```python
some_movies(100)
```
