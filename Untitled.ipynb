```python
import requests
from bs4 import BeautifulSoup
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc
import os
import folium
from folium import plugins
import pandas as pd
```

```python
def movie_crawling(data):
    for i in range(100):
        url="https://movie.naver.com/movie/point/af/list.nhn?&page="
        url=url+str(i+1)
        req=requests.get(url)
        if req.ok:
            html=req.text
            soup=BeautifulSoup(html, 'html.parser')
            titles=soup.select(".title a.movie")
            points=soup.select(".title em")
            contents=soup.select(".title")
            n=len(titles)
            for i in range(n):
                title=titles[i].get_text()
                point=points[i].get_text()
                contentarr=contents[i].get_text().replace('신고','').split("\n\n")
                content=contentarr[2].replace("\t","").replace("\n","")
                data.append([title,point,content])
```

```python
data=[]
movie_crawling(data)
print(data)
```

```python
movie_df=pd.DataFrame(data, columns=['title','point','content'])
movie_df1=movie_df.head()
movie_df1
```

```python
from konlpy.tag import Okt
from collections import Counter
import pytagcloud
```

```python
def makeGraph():
    font_location = "c:/Windows/fonts/malgun.ttf"
    font_name = font_manager.FontProperties(fname=font_location).get_name()
    rc('font', family=font_name)

    plt.xlabel('영화제목')
    plt.ylabel('평균평점')
    plt.grid(True)    
    plt.bar(range(len(movie_df1.title)), movie_df1.point, align='center')
    plt.xticks(range(len(movie_df1.title)), list(movie_df1.title), rotation='70')

    plt.savefig(os.path.join('data/images/fig01.png'), dpi=300)   
```

```python
makeGraph()
```

```python
def cctv_map():
    popup=[]
    data_lat_log=[]
    df=pd.read_csv("data/cctv/CCTV_20190917.csv",encoding="utf-8")
    for data in df.values:
        if data[4]>0:
            popup.append(data[1])
            data_lat_log.append([data[10],data[11]])
    
    m=folium.Map([35.1803305,129.0516257], zoop_start=11)
    plugins.MarkerCluster(data_lat_log,popups=popup).add_to(m)
    m.save("data/map/map01.html")
```

```python
cctv_map()
```

```python
def saveWordCloud():
    nlp = Okt()
    wordtext=""
    for t in movie_df1.content:
        wordtext+=t+" "
    
#   명사 추출
    nouns = nlp.nouns(wordtext)
#   명사의 갯수  
    count = Counter(nouns)
    
    wordInfo = dict()
#                         빈도수 높은 50개
    for tags, counts in count.most_common(50):
#               한 글자 보다 큰 것
        if (len(str(tags)) > 1):
            wordInfo[tags] = counts
    filename=('data/images/wordcloud01.png')
    taglist = pytagcloud.make_tags(dict(wordInfo).items(), maxsize=80)
    pytagcloud.create_tag_image(taglist, filename, 
                                size=(640, 480), fontname='korean', rectangular=False)  
```

```python
saveWordCloud()
```
