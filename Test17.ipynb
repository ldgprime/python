```python
html="""
<html><body>
<h1></h1>
<div>div1</div>
<div>div2</div>
<div id='main'>
<h1 i></h1>
<ul class='items'>
    <li>  </li>
    <li>HTML5</li>
    <li>Python</li>
</ul>
</div>
</body></html>
"""

soup=BeautifulSoup(html,'html.parser')
h1=soup.select_one('div#main > h1').string
print('h1=',h1)
li_list=soup.select('div#main > ul.items > li')
for li in li_list:
    print('li=',li.string)
```

```python
#   
from bs4 import BeautifulSoup 
import urllib.request 
url='https://ko.wikisource.org/wiki/%EC%A0%80%EC%9E%90:%EC%9C%A4%EB%8F%99%EC%A3%BC'
res=urllib.request.urlopen(url)
soup=BeautifulSoup(res,'html.parser')
li_list=soup.select('.mw-parser-output > ul > li')
for title in li_list:
    print(title.a.string)
    li=title.select('li')    
    if li:
        for i in li:
            print('-',i.a.string)
            print()
            url='https://ko.wikisource.org'+i.a.attrs['href']        
            res=urllib.request.urlopen(url).read()
            soup=BeautifulSoup(res,'html.parser')
            content=soup.select_one('#mw-content-text > div.mw-parser-output p').text           
            print(content)
            print()
            #db, df     
    else:
        url='https://ko.wikisource.org'+title.a.attrs['href']
        res=urllib.request.urlopen(url).read()
        soup=BeautifulSoup(res,'html.parser')
        content=soup.select_one('#mw-content-text > div.mw-parser-output p').text
        print(content)
    
# li_list2=soup.select('.mw-parser-output > ul > li > ul > li')
# for li in li_list2:
# print(li.a.string)
```

```python
from pdfminer.pdfinterp import PDFResourceManager,process_pdf
from pdfminer.converter import TextConverter
from pdfminer.layout import LAParams
from io import StringIO
from io import open
from urllib.request import urlopen
```

```python
def readPDF(pdfFile):
    rsrcmgr=PDFResourceManager() # pdf 
    retstr=StringIO() # pdf     
    laparams=LAParams() #  
    #pdf     
    device=TextConverter(rsrcmgr,retstr,laparams=laparams)
    process_pdf(rsrcmgr,device,pdfFile)
    device.close()
    content=retstr.getvalue()
    retstr.close()
    return content
```

```python
pdfFile=urlopen('http://www.pythonscraping.com/pages/warandpeace/chapter1.pdf')
outputString = readPDF(pdfFile)
pdfFile.close()

with open('data/pdf/result.txt','w') as f:
    f.write(outputString)
    print('')
```
